# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Corn_Yield.csv')
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 2)
X_poly = pf.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y)
```

## Visualising the Linear Regression results

```
plt.scatter(X, Y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("Yeild vs Fertilizer used(Linear Regression)")
plt.xlabel("Fertilizer Used")
plt.ylabel("Corn Yeild")
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, Y, color = 'red')
plt.plot(X, pol_reg.predict(X_poly), color = 'blue')
plt.title("Yeild vs Fertilizer used(Polynomial Regression)")
plt.xlabel("Fertilizer Used")
plt.ylabel("Corn Yeild")
plt.show()
```

## Predicting a new result with Linear Regression

```
lin_reg.predict([[16.5]])
```

## Predicting a new result with Polynomial Regression

```
pol_reg.predict(pf.fit_transform([[16.5]]))
```
